```python
# import pandas as pd

def compute_mean(data):
    
    num_rows = len(data)
    num_cols = len(data.columns)

    # Initialize a list to hold the sum for each dimension
    dimension_sum = [0] * num_cols

    # Calculate the sum for each dimension
    for col in data.columns:
        dimension_sum[col] = data[col].sum()

    # Calculate the mean for each dimension
    mean_values = [sum_value / num_rows for sum_value in dimension_sum]

    return mean_values


data_set = pd.read_csv('C:/Users/akmik/OneDrive/Desktop/CSCI 347/Mini Project 1/wine/wine.data', header=None)

print(compute_mean(data_set))
```

```python
import pandas as pd

def compute_correlation(x, y):
    
    n = len(x)

    # Compute means
    mean_x = sum(x) / n
    mean_y = sum(y) / n

    # Compute standard deviations
    std_dev_x = (sum((xi - mean_x) ** 2 for xi in x) / n) ** 0.5
    std_dev_y = (sum((yi - mean_y) ** 2 for yi in y) / n) ** 0.5

    # Compute correlation
    correlation = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n)) / (n * std_dev_x * std_dev_y)

    return correlation

data_set = pd.read_csv('C:/Users/akmik/OneDrive/Desktop/CSCI 347/Mini Project 1/wine/wine.data', header=None)

# Extract the first and second attributes from the dataset
x = data_set.iloc[:, 0].tolist()  # Assuming the first attribute is in the first column
y = data_set.iloc[:, 1].tolist()  # Assuming the second attribute is in the second column

print(compute_correlation(x, y))
```

```python
import pandas as pd

def range_normalize(data):
    
    num_rows, num_cols = data.shape

    normalized_data = pd.DataFrame()

    for col in data.columns:
        # Find the minimum and maximum values for each column
        min_val = data[col].min()
        max_val = data[col].max()

        # Normalize each value in the column
        normalized_column = (data[col] - min_val) / (max_val - min_val)
        normalized_data[col] = normalized_column

    return normalized_data

data_set = pd.read_csv('C:/Users/akmik/OneDrive/Desktop/CSCI 347/Mini Project 1/wine/wine.data', header=None)

normalized_data_set = range_normalize(data_set)
print(normalized_data_set)
```

```python
import pandas as pd

def label_encode(data):
    
    label_map = {}
    encoded_data = pd.DataFrame()

    # Iterate over each column
    for col in data.columns:
        unique_values = data[col].unique()

        # Create a mapping from unique values to labels
        for i, value in enumerate(unique_values):
            if value not in label_map:
                label_map[value] = i

        # Label encode the data in the current column
        encoded_column = [label_map[value] for value in data[col]]
        encoded_data[col] = encoded_column

    return encoded_data, label_map

data_set = pd.read_csv('C:/Users/akmik/OneDrive/Desktop/CSCI 347/Mini Project 1/wine/wine.data', header=None)

encoded_data, label_map = label_encode(data_set)
print("Encoded data:")
print(encoded_data)
print("\nLabel map:")
print(label_map)
```

```python
import pandas as pd

def compute_correlation_matrix(data):
    
    num_features = data.shape[1]
    num_samples = data.shape[0]

    # Compute means of each feature
    means = data.mean()

    # Compute standard deviations of each feature
    std_devs = data.std()

    # Compute correlation between each pair of features
    correlation_matrix = pd.DataFrame(index=data.columns, columns=data.columns)
    for i in range(num_features):
        for j in range(num_features):
            covariance = ((data.iloc[:, i] - means[i]) * (data.iloc[:, j] - means[j])).sum() / num_samples
            correlation = covariance / (std_devs[i] * std_devs[j])
            correlation_matrix.iloc[i, j] = correlation

    return correlation_matrix

def count_high_correlation_pairs(correlation_matrix, threshold=0.5):
    
    num_features = len(correlation_matrix)
    num_high_correlation_pairs = 0

    for i in range(num_features):
        for j in range(i+1, num_features):  # Avoid counting pairs twice
            if correlation_matrix.iloc[i, j] >= threshold:
                num_high_correlation_pairs += 1

    return num_high_correlation_pairs

data_set = pd.read_csv('C:/Users/akmik/OneDrive/Desktop/CSCI 347/Mini Project 1/wine/wine.data', header=None)

# Compute the correlation matrix
correlation_matrix = compute_correlation_matrix(data_set)

# Count the number of pairs of features with correlation greater than or equal to 0.5
num_pairs = count_high_correlation_pairs(correlation_matrix)
print("Number of pairs of features with correlation greater than or equal to 0.5:", num_pairs)
```

```python
import csv
import matplotlib.pyplot as plt

def read_csv_data(data_path):
    """Reads data from a CSV file."""
    data = []
    with open(data_path, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            data.append(row)
    return data

def z_score_normalize(data):
    """Z-score normalizes the data."""
    normalized_data = []
    num_attributes = len(data[0])
    for i in range(1, num_attributes):  # Skip the first column (categorical variable)
        column_values = [float(row[i]) for row in data]
        mean = sum(column_values) / len(column_values)
        std_dev = (sum((val - mean) ** 2 for val in column_values) / len(column_values)) ** 0.5
        normalized_column = [(val - mean) / std_dev for val in column_values]
        normalized_data.append(normalized_column)
    return list(zip(*normalized_data))

def compute_correlation_matrix(data):
    """Computes the correlation matrix for the data."""
    num_attributes = len(data[0])
    correlation_matrix = [[0] * (num_attributes - 1) for _ in range(num_attributes - 1)]  # Exclude the first column
    num_samples = len(data)
    for i in range(1, num_attributes):  # Skip the first column (categorical variable)
        for j in range(1, num_attributes):  # Skip the first column (categorical variable)
            if i == j:
                correlation_matrix[i-1][j-1] = 1
            elif i < j:
                cov = sum((data[k][i] * data[k][j]) for k in range(num_samples)) / (num_samples - 1)
                correlation_matrix[i-1][j-1] = cov
                correlation_matrix[j-1][i-1] = cov
    return correlation_matrix

def find_smallest_correlation(data_path):
    """Finds which Z-score-normalized numerical attributes have the smallest correlation."""
    data = read_csv_data(data_path)
    normalized_data = z_score_normalize(data)
    correlation_matrix = compute_correlation_matrix(normalized_data)

    min_correlation_value = float('inf')
    min_correlation_indices = None
    num_attributes = len(correlation_matrix)

    for i in range(num_attributes):
        for j in range(i+1, num_attributes):
            if correlation_matrix[i][j] < min_correlation_value:
                min_correlation_value = correlation_matrix[i][j]
                min_correlation_indices = (i+1, j+1)  # Increment by 1 to account for skipped first column

    return min_correlation_indices, min_correlation_value

def plot_scatter(data_path, indices):
    """Plots a scatter plot of Z-score-normalized attributes with the smallest correlation."""
    data = read_csv_data(data_path)
    normalized_data = z_score_normalize(data)
    x_indices, y_indices = indices
    x_data = [row[x_indices] for row in normalized_data]
    y_data = [row[y_indices] for row in normalized_data]

    plt.scatter(x_data, y_data)
    plt.xlabel(f"Attribute {x_indices}")
    plt.ylabel(f"Attribute {y_indices}")
    plt.title(f"Scatter Plot of Z-score-Normalized Attributes {x_indices} vs {y_indices}")
    plt.show()

def main(data_path):
    indices, correlation = find_smallest_correlation(data_path)
    print("Z-score-normalized attributes with the smallest correlation:", indices)
    print("Correlation coefficient:", correlation)
    plot_scatter(data_path, indices)

# Example usage:
data_path = 'C:/Users/akmik/OneDrive/Desktop/CSCI 347/Mini Project 1/wine/wine.data'
main(data_path)
```
